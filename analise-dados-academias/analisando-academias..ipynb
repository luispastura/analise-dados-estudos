```python
import pandas as pd

# Use the raw URL to access the CSV file directly
df = pd.read_csv('c:/Users/Usuário/Documents/Estudos/Python/analises-dados/analise-dados-academias/gym_members_exercise_tracking.csv')

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Calcular a média de peso por idade
media_peso_por_idade = df.groupby('Age')['Weight (kg)'].mean().reset_index()

# Configurar o gráfico
plt.figure(figsize=(12, 6))
sns.barplot(
    data=media_peso_por_idade,
    x='Age',
    y='Weight (kg)',
    palette='viridis', 
    edgecolor='blue'  # Define as bordas das barras como azul
)
plt.title('Média de Peso por Idade', fontsize=16)
plt.xlabel('Idade', fontsize=14)
plt.ylabel('Peso Médio (kg)', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()

# Exibir o gráfico
plt.show()
```
